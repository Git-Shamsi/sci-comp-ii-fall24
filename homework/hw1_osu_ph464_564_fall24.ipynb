5) Compute the mean of $x$ and the mean of $y$. Now compute the median of $x$ and the median of $y$. You should note that the means are different, but the medians (as it is a cumulative statistic) are the same. The mean is affected by the scale of the sample values, but the median only depends on the ordering of the samples. Monotonic transformations (like taking the log) do not change the ordering of samples.

### Solution




```python
import numpy as np
import matplotlib.pyplot as plt

# 1) Draw 1000 random samples from a uniform distribution between 0.1 and 10
x = np.random.uniform(0.1, 10, 1000)

# 2) Use matplotlib to make a histogram of these samples
plt.figure(figsize=(8,6))
plt.hist(x, bins=40, color='skyblue', edgecolor='black')
plt.title('Histogram of 1000 random samples from Uniform Distribution (0.1, 10)')
plt.xlabel('x')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

#3) Compute the base-10 log of your array ùë•, and store this as ùë¶

y = np.log10(x)

#4) Make another histogram for ùë¶

plt.figure(figsize=(8,6))
plt.hist(y, bins=40, color='lightgreen', edgecolor='black')
plt.title('Histogram of log10(x)')
plt.xlabel('y (log10(x))')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# 5) Compute the mean and median of x and y
mean_x = np.mean(x)
mean_y = np.mean(y)
median_x = np.median(x)
median_y = np.median(y)

print(f"mean of x and y is {mean_x, mean_y}, and the median is  {median_x, median_y} ")

print("mean is very diffrent but the median is very close which confirm our hypothesis ")
```

## Problem 3

This question is about IQ. It's a flawed and questionable metric, but useful for this problem. By definition, IQ is calibrated as a Gaussian distribution with $\mu=100$ and $\sigma=15$. 

1) Create a `scipy.stats` Gaussian object with these properties. Plot the distribution, and print out 10 random draws, along with the pdf at $x=145$. 

2) Access the `cdf` property of the Gaussian object and plot it over the same $x$-range as in (1) but in a different figure. 

3) What fraction of people have IQ > 145? *Hint: investigate all the properties of the Gaussian Python object from part (1) to find the relevant property that gives you this, or perform a calculation involving the cdf.*

4) What IQ corresponds to "one in a million"?

### Solution

```python
from scipy.stats import norm

# 1) Create a Gaussian distribution with mean=100 and std=15
mu = 100
sigma = 15
distG = norm(mu, sigma)

# Plot the Gaussian distribution
x_vals = np.linspace(40, 160, 1000)
pdf_vals = distG.pdf(x_vals)

plt.figure(figsize=(8,6))
plt.plot(x_vals, pdf_vals, label='Gaussian Distribution (Œº=100, œÉ=15)', color='blue')
plt.title('IQ Distribution (Gaussian, Œº=100, œÉ=15)')
plt.xlabel('IQ')
plt.ylabel('Probability Density')
plt.grid(True)
plt.show()

# Generate 10 random draws
random_draws = distG.rvs(10)
print(random_draws)
```

```python
# 2) Access the CDF of the Gaussian object and plot it over the same x-range
cdf_vals = distG.cdf(x_vals)

plt.figure(figsize=(8,6))
plt.plot(x_vals, cdf_vals, label='Cumulative Distribution Function (CDF)', color='green')
plt.title('Cumulative Distribution of IQ (Œº=100, œÉ=15)')
plt.xlabel('IQ')
plt.ylabel('Cumulative Probability')
plt.grid(True)
plt.show()

# 3) Fraction of people with IQ > 145
prob_greater_than_145 = 1 - distG.cdf(145)

# 4) What IQ corresponds to "one in a million"?
# "One in a million" means the probability is 1/1,000,000
one_in_a_million_prob = 1 - 1e-6
iq_one_in_a_million = distG.ppf(one_in_a_million_prob)

prob_greater_than_145, iq_one_in_a_million
```
